## 5. Simple business questions

```python
total_players = df['player_id'].nunique()
total_revenue = df['in_app_purchases_gbp'].sum()
day_1_retention = df['retained_day_1'].mean() * 100
day_7_retention = df['retained_day_7'].mean() * 100
day_30_retention = df['retained_day_30'].mean() * 100
churn_rate = df['churned'].mean() * 100

print(f'Total players: {total_players}')
print(f'Total revenue: £{total_revenue:,.2f}')
print(f'Day 1 retention: {day_1_retention:.1f}%')
print(f'Day 7 retention: {day_7_retention:.1f}%')
print(f'Day 30 retention: {day_30_retention:.1f}%')
print(f'Churn rate: {churn_rate:.1f}%')
```

## 6. Players by acquisition channel

```python
df['acquisition_channel'].value_counts()
```

## 7. Retention by acquisition channel

The retention analysis shows that Organic and App Store Search channels produced the strongest long-term player retention. Organic had the highest Day 30 retention at 46.9% and the lowest churn rate at 53.1%, suggesting that players who discover the game naturally may be more genuinely interested and more likely to stay active. App Store Search also performed well, which may indicate strong user intent.

In contrast, TikTok Ads and Influencer Campaigns had lower Day 30 retention and higher churn. This suggests that paid or influencer-led campaigns may attract many players initially, but some of those players may not remain engaged over time. From a CRM and user acquisition perspective, the company should not only focus on install volume but also on retention quality.

```python
retention_by_channel = df.groupby('acquisition_channel')[['retained_day_1', 'retained_day_7', 'retained_day_30', 'churned']].mean() * 100
retention_by_channel.round(1)
```
